```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 준비
tips = sns.load_dataset("tips")
tips_gr = tips.groupby('day').count()
tips_gr

# barplot 그리기
sns.barplot(x = tips_gr.index, y = tips_gr['tip'],  palette='YlGnBu')
plt.show()

# barhplot 그리기
sns.barplot(x = tips_gr['tip'], y = tips_gr.index,  palette='YlGnBu')
plt.show()
```

```python
# Boxplot : 중앙값이 비교가 됨
# 전체 데이터의 50%가 박스 안에 분포되어있음
# 박스의 높이가 높으면 값이 많이 분산되어 있다(분포가 크다).(박스의 높이가 낮으면 값의 집중되어 있다.)
# 선은 이상값의 경계를 표현한 것이다.
# 선 길이 내에 들어오면 이 안에 있는 값이 정상적인 값, 선 밖에 있으면 비정상적인 값

import seaborn as sns
import matplotlib.pyplot as plt

# seaborn 내장 데이터셋 load_dataset을 사용하여 'tips' 데이터프레임 불러오기
tips = sns.load_dataset('tips')

# 박스플롯 그리기
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips)
plt.show()
```

```python
# 중앙값 : 전체 데이터를 크기순으로 정렬했을 때, 전체 데이터의 갯수의 절반에 위치한 값이 중앙값
# 데이터의 갯수가 홀수 개면 가운데 있는 값이 중앙값, 데이터의 갯수가 짝수 개면 가운데의 있는 두값을 더해 2로 나눈 값

# 사분위수 (데이터의 갯수를 정확하게 4등분한 것)
# 전체를 크기순으로 나열하여 25% 지점에 있는 값 : Q1
# 전체를 크기순으로 나열하여 50% 지점에 있는 값 : Q2 => 중앙값
# 전체를 크기순으로 나열하여 75% 지점에 있는 값 : Q3
# Q3 - Q1 = 50%(IQR) : 내 데이터에 대한 범위
# 이상값의 범위를 구하는 방법은 Q1 - 1.5 * IQR이다.(왼쪽 이상값 범위 구하는 방법)
# Q3 + 1.5 * IQR  (오른쪽 이상값 범위 구하는 방법)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# seaborn 내장 데이터셋 load_dataset을 사용하여 'tips' 데이터프레임 불러오기
tips = sns.load_dataset('tips')

# 박스플롯 그리기
sns.boxplot(x='day', y='total_bill', data=tips, palette=['blue', 'orange', 'green', 'red'])
plt.show()
```

```python
# 기본 joint plot
sns.jointplot(x="total_bill", y="tip", height=7, data=tips)
plt.show()
```

```python
# 회귀선 추가
sns.jointplot(x="total_bill", y="tip", height=7, data=tips, kind="reg")
plt.show()
```

```python
# hex모양의 밀도추가
sns.jointplot(x="total_bill", y="tip", height=7, data=tips, kind="hex")
plt.show()
```

```python
# 등고선모양의 밀도추가
iris = sns.load_dataset('iris')
sns.jointplot(x="sepal_width", y="petal_length", height=7, data=iris, kind="kde", color="g")
plt.show()
```

```python
# 사분위수 찾기
q2 = df.tip.median()
# 크기순으로 나열
df_sr = df.tip.sort_values().reset_index(drop=True)
df_sr
# 전체 개수중 가운데 위치값
q2_loc = len(df_sr) / 2
# q2_loc로 df_sr 인덱싱
df_sr[q2_loc] # 이렇게 찾은 2.92가 중앙값
# q1
q1_loc = len(df_sr) * 1/4
df_sr[q1_loc] # 이렇게 찾은 2.92가 중앙값
# q3
q3_loc = len(df_sr) * 3/4
df_sr[q3_loc]
```

```python
q1 = df['tip'].quantile(0.25)
q2 = df['tip'].quantile(0.5)
q3 = df['tip'].quantile(0.75)
```

```python
# IQR
iqr = q3 - q1
iqr
```

```python
# outlier
out_min = q1 - (1.5 * iqr)
out_max = q3 + (1.5 * iqr)
print(out_min, out_max)
```
