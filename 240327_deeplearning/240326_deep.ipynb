```python
# AND 게이트
def AND(a, b):
    x1, x2 = (a, b)
    w1, w2 = (-0.5, -1)
    bias = 4
    y = (w1 * x1) + (w2 * x2) + bias
    if y >= 3:
        return 0
    else:
        return 1
```

```python
# 이제부터 학습시작
# 입력값은 0 또는 1
print(AND(0,0)) # 0
print(AND(0,1)) # 0
print(AND(1,0)) # 0
print(AND(1,1)) # 1
```

```python
# OR 게이트
def OR(a, b):
    x1, x2 = (a, b)
    w1, w2 = (1.5, 1.5)
    bias = -1
    y = (w1 * x1) + (w2 * x2) + bias
    if y <= 0: 
        return 0
    else:
        return 1
```

```python
print(OR(0,0)) # 0
print(OR(0,1)) # 1
print(OR(1,0)) # 1
print(OR(1,1)) # 1
```

```python
# NAND 게이트
def NAND(a, b):
    x1, x2 = (a,b)
    w1, w2 = (-0.5,-0.7)
    bias = 1
    y = (w1 * x1) + (w2 * x2) + bias
    if y <= 0:
        return 0
    else: 
        return 1
```

```python
print(NAND(0,0)) # 1
print(NAND(0,1)) # 1
print(NAND(1,0)) # 1
print(NAND(1,1)) # 0
```

```python
def XOR(a, b):
    i1 = NAND(a, b)
    i2 = OR(a, b)
    y = NAND(i1, i2)
    return y
```

```python
print(XOR(0, 0)) # 0
print(XOR(0, 1)) # 1
print(XOR(1, 0)) # 1
print(XOR(1, 1)) # 0
```

```python
# 시그모이드 함수 만들기
import numpy as np
import matplotlib.pyplot as plt

# linspace : 시작점에서 끝점까지를 등분수만큼 등분하는 함수
x = np.linspace(-10, 10, 100)
y = 1/(1 + np.exp(-x)) # exp는 자연상수

plt.plot(x, y, 'or')
plt.show()
```

```python
# 하이퍼볼릭 탄젠트 함수 만들기
x = np.linspace(-10, 10, 100)
y = np.tanh(x)

plt.plot(x, y, 'or')
plt.show()
```

```python
# ReLu 함수
def ReLu(x):
    arr = []
    for a in x:
        if a > 0:
            arr.append(a)
        else:
            arr.append(0)
    return np.array(arr)

x = np.linspace(-10, 10, 100)
y = ReLu(x)
plt.plot(x, y, 'or')
plt.show()
```

```python
# Leaky ReLu 함수
def LeakyReLu(x):
    arr = []
    for a in x:
        if a <= 0:
            arr.append(a * 0.1)
        else:
            arr.append(a)
    return np.array(arr)

x = np.linspace(-10, 10, 100)
y = LeakyReLu(x)
plt.plot(x, y, 'or')
plt.show()
```

```python
# Softmax 함수 : 입력값을 각 클래스에 대한 확률값으로 변환하여 출력
x = np.linspace(-10, 10, 100)
y = np.exp(x) / np.sum(np.exp(x))

sum(y) # y값 다 더하면 1 => 확률값
```

```python
# Softmax 함수
x = np.linspace(-3, 7, 3)
y = np.exp(x) / np.sum(np.exp(x))

y
np.where(y.max() == y) # array([2], dtype=int64) 최댓값의 인덱스 주소
```
