```python
# 막대그래프는 범주형 데이터의 크기를 나타내는 그래프, 막대그래프에 대한 크기 차이가 별로 안나면 효과가 없다.
# y축이 숫자가 될수 있는 데이터면 뭐든 괜찮다. x축은 범주형 데이터만 올 수 있다.
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
values2 = [150, 300, 850]
colors = ['y', 'dodgerblue', 'C2']

plt.bar(x-0.2, values, color=colors, width=0.4) # 막대그래프는 x축과 y축을 같이 입력해줘야함
plt.bar(x+0.2, values2, width=0.4) # width의 절반의 간격을 x축에 주면 된다.
plt.xticks(x, years)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

ids = pd.Series(['m01', 'm02', 'm03', 'm04'])
before = pd.Series([100, 110, 150, 90])
after = pd.Series([160, 130, 170, 120])
df = pd.concat([ids, before, after], axis=1)
df.columns = ['ID', 'Before', 'After']

df
```

```python
fig = plt.Figure()
axes1 = plt.subplot(2,1,1)
axes2 = plt.subplot(2,1,2)

plt.suptitle('Squat')

axes1.set_title('Before')
axes2.set_title('After')
axes1.set_xlabel('ID')
axes1.set_ylabel('Weight')

axes1.bar(df['ID'],df['Before'], width=0.8, color=['r', 'g', 'b', 'm'])

# 수평 그래프는 barh로 생성
axes2.barh(df['ID'],df['After'], color=['r', 'g', 'b', 'm'])
# ID 순서를 바꾸는 방법
# axes2.barh(df['ID'].sort_index(ascending=False),df['After'], color=['r', 'g', 'b', 'm'])

plt.subplots_adjust(hspace=0.8) # 그래프 사이의 간격 조정
plt.show()
```

```python
plt.bar(df.index, df['Before'], width=0.4, color='yellow', label='Before', align='edge')
plt.bar(df.index+0.4, df['After'], width=0.4, color='orange', label='After', align='edge')

plt.title('Squat')
plt.xlabel('ID')
plt.ylabel('Weight')

plt.xticks(x, df['ID'])

plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='winter')
plt.colorbar()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dic = {'city': ['Seoul','Incheon','Dejeon','Degu','Ulsan','Busan','Kawngju'],
       'lat':[37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16],
       'lon':[126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85],
       'p_dense': [16154, 2751, 2839, 2790, 1099, 4454, 2995]}
df = pd.DataFrame(dic)
df
plt.scatter(df['lon'], df['lat'], s=df['p_dense']*0.2, c=df.index, alpha=0.5, cmap='spring')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Dense')

# text 입력
for i in df.index:
    plt.text(df['lon'][i], df['lat'][i], df['city'][i])

plt.colorbar()
plt.show()
```
