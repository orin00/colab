```python
import pandas as pd
import numpy as np
import seaborn as sns

df = sns.load_dataset('tips')
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
## tip 예측 모델
# 종속변수와 독립변수를 나눠줘야 한다.
## y: tip,  X: 나머지
y = df.pop('tip')
X = df
```

```python
y
```

```python
X
```

```python
## 모델 불러오기
from sklearn.linear_model import LinearRegression
lrm = LinearRegression()
```

```python
# 글자를 숫자로 바꾸기(인코딩)
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

encoder = LabelEncoder()
cols = ['sex', 'smoker','day', 'time']
X_cat = pd.DataFrame()
# LabelEncoder() 라는 친구는 칼럼 하나씩만 변형시킬 수 있다.
for i in cols:
  X_cat[i] = encoder.fit_transform(X[i])

X_cat
```

```python
## 숫자 데이터의 크기 조정 (Scaling)
scaler = MinMaxScaler()

X_num = scaler.fit_transform(X[['total_bill', 'size']])
X_num = pd.DataFrame(X_num,columns = ['total_bill', 'size'])
X_num
```

```python
## 다시 하나로 합치기
X = pd.concat([X_cat, X_num],axis=1)
```

```python
# 학습 데이터와 평가 데이터 분리
# test_size는 0.25가 기본값
# random_state는 고정된 값에 의해서 지정된 랜덤 값만 뽑힘
from sklearn.model_selection import train_test_split
(X_train, X_test, y_train, y_test) = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
print(X_train.shape, X_test.shape) # X, y끼리는 열이 맞는지 보고
print(y_train.shape, y_test.shape) # train끼리 test끼리 행이 맞는지 봐야함
```

```python
## 모델 학습
lrm.fit(X_train, y_train)
```

```python
pred = lrm.predict(X_test) # 예측값 생성
pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score
from math import sqrt

print('MAE : ', mean_absolute_error(y_test, pred))
print('MSE : ', mean_squared_error(y_test, pred))
print('RMSE : ', sqrt(mean_squared_error(y_test, pred)))
print('MAPE : ', mean_absolute_percentage_error(y_test, pred))
print('r2 Score : ', r2_score(y_test, pred))
```
