```python
import pandas as pd

data = {'숫자': [1,2,3],
        '문자': ['a','b','c'],
        '한글': ['가','나','다']}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.rename(index={0:100, 1:200, 2:300})
df
```

```python
df.rename(index={0:100, 1:200, 2:300}, columns={'숫자':'num', '문자':'str', '한글':'kor'})
df
```

```python
df.columns= ['num','str','kor']
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
idx = ["037730", "036360", "005760"]
col = ['종목명', '현재가']
data = [["3R", "1510"],
         ["3SOFT", 1790],
         ["ACTS", 1185]]
df = pd.DataFrame(data, columns=col, index=idx)
df
```

```python
data = [3300, 2000, 5000]
date = ['2/26', '2/27', '2/28']

sr_data = pd.Series(data)
sr_date = pd.Series(date)
```

```python
sr_data
```

```python
sr_date
```

```python
df = pd.DataFrame([sr_data, sr_date])
df
```

```python
# 첫번째 리스트 안에 요소 : row(행), 각 요소 안에 요소 : column(열)
# 이를 해결하는 방법
# 딕셔너리로 만들면 된다.
df = pd.DataFrame({'data': sr_data,
                   'date': sr_date})
df
```
