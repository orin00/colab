```python
import cv2
import numpy as np

img = cv2.imread('./images/fish.jpg')
img = cv2.resize(img, (680, 420))
img.shape # row, column, channel
```

```python
# 원근법만 필요하므로 채널 정보는 필요없음
rows, cols = img.shape[:-1]
```

```python
pts1 = np.float32([[0,0],
                   [0,rows],
                   [cols,0],
                   [cols,rows]])

pts2 = np.array([[100,50],
                   [10,rows-50],
                   [cols-100,50],
                   [cols-10,rows-50]], dtype='float32')
```

```python
cv2.imshow('IMG', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 끝 점 표시
cv2.circle(img, (0,0), 10, (255,0,0), -1)
cv2.circle(img, (0, rows), 10, (0,255,0), -1)
cv2.circle(img, (cols, 0), 10, (0,0,255), -1)
cv2.circle(img, (cols, rows), 10, (0,255,255), -1)
```

```python
#getPerspectiveTransform는 행렬 만들어주는애
M = cv2.getPerspectiveTransform(pts1, pts2)
```

```python
dst = cv2.warpPerspective(img, M, (cols, rows))
cv2.imshow('IMG', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

title = "Scanner"
img = cv2.imread("./images/paper.jpg")

# 2차원으로 원근 적용할 전체 크기
rows, cols = img.shape[:-1]
img2 = img.copy() # 원근 변환이 적용될 이미지
pts_cnt = 0 # 필요한 4개의 포인트인지 확인하기 위한 변수
pts = np.zeros((4,2), dtype=np.float32) # 마우스로 클릭할 위치에 대한 변환

# 마우스 클릭에 사용할 콜백 함수
def onMouse(event, x, y, flags, param):
    global pts_cnt
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img2, (x, y), 10, (0,255,0), -1) # 클릭시 이미지에 원그리기
        cv2.imshow(title, img2)

        pts[pts_cnt] = [x,y] # 마우스 버튼 좌표 저장
        pts_cnt += 1 # 마우스 버튼 클릭 횟수 업데이트
        print(x, y)
        if pts_cnt == 4:
            # 각 좌표의 위치(상하좌우)를 찾는 코드
            sm = pts.sum(axis=1) # x + y

            # diff는 차분이라는 게 있다. 시계열 데이터를 사용하기 때문에 순차적이다.
            # 차분 : 앞에 있는 데이터 값이 뒤에 있는 데이터 값에 영향을 준다.
            # 그래서 diff = np.diff(pts, axis=1) 이 부분은 y - x 로 적용되었다.
            diff = pts[:, 0] - pts[:, 1] # 이렇게 하면 x - y로 적용
            
            print(pts)
            
            top_left = pts[np.argmin(sm)] # 좌측 상단
            top_right = pts[np.argmax(diff)] # 우측 상단
            bottom_left = pts[np.argmin(diff)] # 좌측 하단
            bottom_right = pts[np.argmax(sm)] # 우측 하단
            
            pts1 = np.float32([top_left, top_right, bottom_right, bottom_left])

            print(pts1)
            
            # 각 꼭지점들의 거리 계산
            w1 = abs(bottom_left[0] - bottom_right[0])
            w2 = abs(top_left[0] - top_right[0])
            h1 = abs(top_right[1] - bottom_right[1])
            h2 = abs(top_left[1] - bottom_left[1])
            width = int(max([w1, w2]))
            height = int(max([h1, h2]))

            # 변환하는 위치의 값
            pts2 = np.float32([[0, 0], [width-1, 0],
                               [width-1, height-1], [0, height-1]])
            print(pts2)
            
            # 변환 행렬 계산
            M = cv2.getPerspectiveTransform(pts1, pts2)

            # 변환 행렬 적용
            res = cv2.warpPerspective(img, M, (width, height))
            cv2.imshow('Result', res)

cv2.imshow(title, img)
cv2.setMouseCallback(title, onMouse)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

kernel_size = (20,20) # 블러 처리에 사용할 커널 크기
title = 'mosaic'
img = cv2.imread('images/self.jpg')
img = cv2.resize(img, (720, 960))

while True:
    # 스페이스바를 누르면 선택영역에 블러가 적용된다.
    x,y,w,h = cv2.selectROI(title, img, False)
    if w > 0 and h > 0:
        roi = img[y:y+h, x:x+h]
        roi = cv2.blur(roi, kernel_size)
        img[y:y+h, x:x+h] = roi
        cv2.imshow(title, img)
    else:
        break
# while 반복문으로 인해 waitKey를 쓸 필요가 없다.
#스페이스바 한번 더 누르면 종료됨
cv2.destroyAllWindows()
```

```python
# 웹캠 있어야 가능
cap = cv2.VideoCapture(0)

if cap.isOpened():
    while True:
        ret, img = cap.read()
        if ret:
            x,y,w,h = ((img.shape[0]//2-50), (img.shape[1]//2-50), 50, 50)
            roi = img[y:y+h, x:x+h]
            roi = cv2.blur(roi, kernel_size)
            img[y:y+h, x:x+h] = roi
            cv2.imshow(title, img)
            if cv2.waitKey(1) != -1:
                break
        else:
            break
cap.release()                
cv2.destroyAllWindows()
```

```python
# 경계 검출
import cv2
import numpy as np

img = cv2.imread('images/sudoku.jpg')
img = cv2.resize(img, (600,600))

gx_kernel = np.array([[-1,1]])
gy_kernel = np.array([[-1],
                      [1]])

# -1은 img의 기본 사이즈 사용
edge_gx = cv2.filter2D(img, -1, gx_kernel) * 2
edge_gy = cv2.filter2D(img, -1, gy_kernel) * 2

merge_img = np.hstack((img, edge_gx, edge_gy))
cv2.imshow('EDGE', merge_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 웹캠에서 필터 적용하는 방법
import cv2
import numpy as np

img = cv2.imread('images/cat1.jpg')
# cap = cv2.VideoCapture(0)
# while cap.isOpened():
#     # 프레임 읽기
#     ret, frame = cap.read()

#     # 속도 향상을 위해 영상크기를 절반으로 줄임
#     fram = cv2.resize(fram, None, fx=0.5, fy=0.5, interpolation = cv2.INTER_AREA)
#     if cv2.waitKey(1) == 27: # esc키로 종료
#         break

# 그레이 스케일로 변경
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 잡음제거를 위해 가우시안 블러 필터 적용(정규 분포를 따라서 블러 처리)
# 이미지를 명확하게 해주기 위해 사용
img_gray = cv2.GaussianBlur(img_gray, (9,9), 0)
# 라플라시안 필터로 엣지 검출
edge = cv2.Laplacian(img_gray, -1, None, 5)
# 스레시홀드로 경계값만 남김
ret, sketch = cv2.threshold(edge, 70, 255, cv2.THRESH_BINARY_INV)

# 경계선 강조하기 위한 값의 팽창 연산
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (3,3))
sketch = cv2.erode(sketch, kernel)
# 경계선 자연스럽게 하기 위해 미디언 블러 필터 적용
sketch = cv2.medianBlur(sketch, 5)
# 그레이 스케일에서 BGR 컬러 스케일로 변경
img_sketch = cv2.cvtColor(sketch, cv2.COLOR_GRAY2BGR)

# 컬러이미지 선명선을 없애기 위해 평균 블러 필터 적용
img_paint = cv2.blur(img, (10,10))

# 컬러 영상과 스케치 영상과 합성
img_paint = cv2.bitwise_and(img_paint, img_paint, mask=sketch)

# 결과 출력
merge_img = np.hstack((img_sketch, img_paint))
cv2.imshow("IMAGE", merge_img)

# cap.release()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
