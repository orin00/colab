```python
import pandas as pd

data = {'숫자': [1,2,3],
        '문자': ['a','b','c'],
        '한글': ['가','나','다']}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.rename(index={0:100, 1:200, 2:300})
df
```

```python
df.rename(index={0:100, 1:200, 2:300}, columns={'숫자':'num', '문자':'str', '한글':'kor'})
```
