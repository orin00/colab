```python
# date 타입으로 바꾸기
X_cat['release_date'] = pd.to_datetime(X_cat['release_date'])
```

```python
# date 타입에서 년, 월 만 뽑음
X_cat['year'] = X_cat['release_date'].dt.year
X_cat['month'] = X_cat['release_date'].dt.month
```

```python
# 날짜 오브젝트 컬럼 삭제
X_cat.drop('release_date', axis =1, inplace=True)
```

장르별로 스플릿

```python
# 장르 별로 ', '단위로 스플릿
X_cat['genres'] = X_cat['genres'].str.split(', ')
```

```python
# 스플릿한 X_cat['genres']에서 유니크한 값들을 뽑아서 리스트에 담기
genres = []
for g in X_cat['genres']:
  for t in g:
    if t not in genres:
      genres.append(t)

genres
```

```python
# 장르값들 만큼의 True False 나눠주는 코드
genres_col = []

for x in X_cat['genres']:
  result = []
  for g in genres:
    if g in x:
      result.append(True)
    else:
      result.append(False)
  genres_col.append(result)
```

```python
# 위에서 나눠준 True False 데이터 프레임으로 나눠주는 코드
df_genres = pd.DataFrame(genres_col, columns=genres)
```

```python
# index를 X_cat의 인덱스로 바꿔주기
df_genres.set_index(X_cat.index, drop=True, inplace=True)
df_genres
```

```python
# X_cat과 df_genres 데이터프레임 합치기
X_cat = pd.concat([X_cat, df_genres], axis=1)
```

```python
# X_cat 데이터프레임 확인
X_cat.head()
```

언어별로 스플릿

```python
# 언어별로 ', '단위로 스플릿
X_cat['spoken_languages'] = X_cat['spoken_languages'].str.split(', ')
```

```python
# 스플릿한 X_cat['spoken_languages']에서 유니크한 값들을 뽑아서 리스트에 담기
language = []
for g in X_cat['spoken_languages']:
  for l in g:
    if l not in language:
      language.append(l)

language
```

```python
# 언어값들 만큼의 True False 나눠주는 코드
language_col = []

for x in X_cat['spoken_languages']:
  result = []
  for g in language:
    if g in x:
      result.append(True)
    else:
      result.append(False)
  language_col.append(result)
```

```python
# 위에서 나눠준 True False 데이터 프레임으로 나눠주는 코드
df_language = pd.DataFrame(language_col, columns=language)
```

```python
# encoding
# label encoding은 유니크 값에다가 숫자를 붙여주는 것(라벨의 숫자에 크기가 적용되어 숫자가 크면 중요도가 높아지거나 하는 문제가 생긴다.)
# one_hot encoding은 유니크 값들을 칼럼으로 만들어서 True False로 values를 변경하는 것을 말함.

one_hot = pd.get_dummies(X_cat['original_language'], drop_first=True)
one_hot
```

```python
X_cat.drop('original_language', axis=1, inplace=True)
df_cat = pd.concat([X_cat, one_hot], axis=1)
df_cat
```

```python
# 학습 데이터와 평가 데이터 분리
from sklearn.model_selection import train_test_split

(X_train, X_test, y_train, y_test) = train_test_split(X, y, test_size=0.3, random_state=4)
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lrm = LinearRegression()
lrm.fit(X_train, y_train)
```

```python
pred = lrm.predict(X_test)
pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score
from math import sqrt

print('MAE : ', mean_absolute_error(y_test, pred))
print('MSE : ', mean_squared_error(y_test, pred))
print('RMSE : ', sqrt(mean_squared_error(y_test, pred)))
print('MAPE : ', mean_absolute_percentage_error(y_test, pred))
print('r2 Score : ', r2_score(y_test, pred))
```
