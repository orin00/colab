```python
# 데이터  변경, 데이터의 그룹화, 데이터프레임의 병합, 데이터프레임 실습

import pandas as pd
import numpy as np

df = pd.DataFrame({'date' :
		       	['01-02-23','02-01-23','12-12-22'],
									'date1':
		       	['01/02/23','02/01/23','12/12/22'],
                  'date2':
		       	['01.02.23','02.01.23','12.12.22']})
df
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%m-%d-%y')
df.dtypes
```

```python
# format은 바꾸려는 부분의 형식 구분을 따라줘야 함
df['date1'] = pd.to_datetime(df['date1'], format='%y/%m/%d')
df
```

```python
df['date3']=["23년2월26일","23년2월27일","23년2월28일"]
df
```

```python
# 년월일로 구분된 데이터를 날짜 데이터로 변형
pd.to_datetime(df['date3'], format="%y년%m월%d일")
```

```python
df['date'][0].dayofweek
```

```python
# 특정 요일의 백화점 매출 현황 같은 정보 조회할 때
df['month'] = df['date'].dt.month
df
```
