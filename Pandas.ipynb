```python
data = np.linspace(1,10,4)
data
```

```python
# 1. 시리즈로 만들기
sr_data = pd.Series(data)
sr_data
```

```python
# 2. 인덱스를 마지막 데이터의 인덱스만 30으로 바꾸기
sr_data.rename({3: 30}, inplace=True)
sr_data
```

```python
# 3. 모든 인덱스를 100으로 바꾸기
sr_data.index = [100]*4
sr_data
```

```python
# 4. 인덱스를 다시 0부터 시작하는 숫자로 바꾸기
sr_data.reset_index(drop=True, inplace=True)
sr_data
```
