```python
group_year_df = df.groupby('year')
print(type(group_year_df)) # DataFrameGroupBy <= 이거 확인 할 수 있으면 됨. 그룹화한 데이터프레임
group_year_df
```

```python
grouped_year_df_lifeExp = group_year_df['lifeExp']
print(type(grouped_year_df_lifeExp)) # SeriesGroupBy <= 이거 확인 할 수 있으면 됨. 그룹화한 시리즈
grouped_year_df_lifeExp
```

```python
mean_lifeExp_by_year = grouped_year_df_lifeExp.mean()
mean_lifeExp_by_year
```

```python
# 공통으로 존재하는 'year', 'continent' 칼럼으로 그룹화해서 'lifeExp', 'gdpPercap' 칼럼을 추출해서 .mean() 평균값 출력
multi_group_var = df.groupby(['year','continent'])[['lifeExp', 'gdpPercap']].mean()
multi_group_var
```

```python
from collections import OrderedDict

# 데이터프레임을 만들 때 순서가 보장된 딕셔너리를 전달하려면 OrderedDict 사용해야 한다.
scientists = pd.DataFrame(OrderedDict([
    ('Name', ['Rosaline Franklin', 'William Gosset']),
    ('Occupation', ['Chemist', 'Statistician']),
    ('Born', ['1920-07-25', '1876-06-13']),
    ('Died', ['1958-04-16', '1937-10-16']),
    ('Age', [37, 61])
  ])
)

scientists
```

```python
scientists = pd.DataFrame(
    data = {'Occupation' : ['Chemist', 'Statistician'],
    'Born' : ['1920-07-25', '1876-06-13'],
    'Died' : ['1958-04-16', '1937-10-16'],
    'Age' : [37, 61]},
    index = ['Rosaline Franklin', 'William Gosset'],
    columns=['Occupation', 'Born', 'Age', 'Died'])

# 데이터프레임에서 시리즈를 선택 loc 속성에 인덱스(과학자 이름)을 전달
first_row = scientists.loc['William Gosset']
print(type(first_row))
```

```python
first_row.index # 시리즈의 인덱스가 들어있다, .index는 속성
```

```python
first_row.index[0] # .index 속성의 0번 인덱스에 있는 값 추출
```

```python
first_row.values # 시리즈의 데이터가 들어있다, .values는 속성
```

```python
first_row.values[0] # .values 속성의 0번 인덱스에 있는 값 추출
```

```python
first_row.keys() # keys는 속성이 아니라 메서드이다. = .index와 같은 기능을 함
```

```python
first_row.keys()[0] # .keys() 메서드의 0번 인덱스에 있는 값 추출
```
