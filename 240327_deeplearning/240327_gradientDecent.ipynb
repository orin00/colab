```python
# 경사하강법을 만들어 보자

import numpy as np
import matplotlib.pyplot as plt

# Learning Rate (학습률)
lr_list = [0.001, 0.1, 0.3, 0.7]

def get_derivative(lr_list):
    # 초기 가중치 임의 설정 (컴퓨터가 초기에 임의의 가중치를 넣는 것 같이)
    weight_old = 2
    # 가중치 목록
    derivative = [weight_old]

    # 손실 함수 정의 : y = x^2(실제 손실함수 x, 계산하기 편하라고)
    y = [weight_old ** 2]

    # epoch = 10 : 반복학습 횟수
    for i in range(1, 10):
        # 손실함수에 대한 기울기 (손실함수 미분)
        dev_value = 2 * weight_old

        ## 가중치 업데이트
        weight_new = weight_old - lr * dev_value
        weight_old = weight_new

        ## 업데이트한 값을 저장
        ## 가중치 업데이트 저장
        derivative.append(weight_old)

        ## 손실값 업데이트 저장
        y.append(weight_old ** 2)

    return derivative, y

## 데이터 만들기
x = np.linspace(-2, 2, 50)
x_square = [i**2 for i in x]

## 그래프 그리기
fig = plt.figure(figsize=(12,7))

# enumerate는 값과 인덱스 값 출력
for i, lr in enumerate(lr_list):
    derivative, y = get_derivative(lr)
    ax = fig.add_subplot(2, 2, i+1)
    ax.scatter(derivative, y, color='red')
    ax.plot(x, x_square)
    ax.title.set_text('Learning Rate = '+ str(lr))
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris['data']
```

```python
y = iris['target']
```

```python
## iris 데이터 딥러닝 모델 만들기
import tensorflow as tf
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y,
                                                    test_size=0.3,
                                                    random_state=1)
```

```python
X_train.shape
```

```python
# 인코딩은 무조건 모델 생성 전에 해준다.
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)
```

```python
## 모델 Sequential 생성
model = tf.keras.Sequential()
## Input Layer 생성
model.add(tf.keras.layers.Flatten(input_shape=(4,)))
## Hidden Layer 생성
model.add(tf.keras.layers.Dense(256, activation='relu'))
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='relu'))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(8, activation='relu'))
## Output Layer 생성
# y의 클래스가 3개 있으므로 3으로 설정
model.add(tf.keras.layers.Dense(3, activation='softmax'))
```

```python
# Model Compile
model.compile(# 경사하강법
              optimizer = 'nadam',
              # 소실함수 (연속형 숫자일때 회귀할떄: MSE, 
              # 분류할때: CategoricalCrossentropy)
              loss=tf.keras.losses.CategoricalCrossentropy(),
              # 평가지표
              metrics=['accuracy'])
```

```python
## 모델 함수 
# epochs가 너무 많으면 과적합 걸린다.
history = model.fit(X_train, y_train, epochs=40)
```

```python
## 내 모델 평가
model.evaluate(X_test, y_test)
```

```python
## 평가된 지표 확인
dict(zip(model.metrics_names, model.evaluate(X_test, y_test)))
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Loss')
plt.plot(history.history['accuracy'], label='ACC')
plt.xlim(-1, 40)
plt.title('Loss')
plt.legend()
plt.show()

# 경사하강법을 모든 데이터에 적용하기 때문에 가중치가 오르락내리락해서
# 그래프가 구불구불해진다.
```
