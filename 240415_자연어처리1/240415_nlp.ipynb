```python
# 형태소만 토크나이징
mor_txt = okt.morphs(text)
print(mor_txt)
```

```python
dic = {}
sorted(mor_txt)
for t in mor_txt:
    if t not in dic.keys():
        dic[t] = 1
    else:
        dic[t] += 1
dic
```

```python
# collections를 모르면 pandas를 사용해서 빈도수 가장 많은 단어 찾기
import pandas as pd

df = pd.DataFrame(dic.items())
df
```

```python
df.sort_values(1, ascending=False)
```

```python
# collections를 모르면 numpy를 사용해서 빈도수 가장 많은 단어 찾기
import numpy as np

uni_txt = np.unique(mor_txt, return_counts=True)
uni_txt[0] ## 중복되지 않은 형태소
uni_txt[1] ## 각 형태소의 빈도수
max_idx = np.argmax(uni_txt[1]) ## 빈도수가 가장 큰값의 인덱스값
uni_txt[0][max_idx]
```

```python
## 2021년 신년사로 워드 클라우드 만들기
from konlpy.tag import Okt
# 형태소 분석기 선택
okt = Okt()

# 신년사 파일 열기
f = open('2021_newyear.txt', 'r', encoding='utf-8')
text = f.read()
f.close()
```

```python
## 신년사에서 명사만 추출하기
nouns = okt.nouns(text)
## 전체 단어의 갯수 100개 선택
from collections import Counter
top_100 = Counter(nouns).most_common(100)
top_100
```

```python
## 이미지 불러오기
from PIL import Image
img = Image.open('mask.png')
mask_arr = np.array(img)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

plt.rcParams['font.family']='Hoon'
wc = WordCloud(font_path='C:\Windows\Fonts\MH.TTF', 
               mask=mask_arr,
               background_color='white',
               colormap='winter')

wordcl = wc.generate_from_frequencies(dict(top_100))
plt.figure(figsize=(10,7))
plt.imshow(wordcl, interpolation='bilinear')
plt.axis('off')
plt.savefig('wordcloud.png', dpi=300, bbox_inches='tight', transparent=True)
```

```python
## BOW(Bag of Words)
txt = ['파이썬 차트 파이썬 머신러닝 python', # 문장 1
       '차트 파이썬 RR 차트', # 문장 2
       'RR 분석 시각화 python'] # 문장 3
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vec = CountVectorizer()
vec.fit_transform(txt).toarray()
```

```python
vec.vocabulary_ # vec.fit_transform(txt).toarray()에서 표시된 단어들의 인덱스 주소값
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vec = CountVectorizer(stop_words=['차트', 'rr']) # 불필요한 단어 전처리
vec.fit_transform(txt).toarray()
```

```python
vec.vocabulary_
```

```python
vec.get_feature_names_out()
```

```python
import pandas as pd

df = pd.DataFrame(vec.fit_transform(txt).toarray())
df.columns = vec.get_feature_names_out()
df
```

```python
# Document-Term Matrix(DTM)
txt = [
	['먹고 싶은 사과'], # doc1
	['먹고 싶은 바나나'], # doc2
	['길고 노란 바나나 바나나'], # doc3
	['저는 과일이 좋아요'] # doc4
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
import pandas as pd

vec = CountVectorizer()
one_list = sum(txt, [])
one_list
values = vec.fit_transform(one_list).toarray()
df = pd.DataFrame(values, columns=vec.get_feature_names_out())
df
```

```python
one_list2 = np.concatenate(txt)
print(one_list2)
values = vec.fit_transform(one_list2).toarray()
df = pd.DataFrame(values, columns=vec.get_feature_names_out())
df
```
